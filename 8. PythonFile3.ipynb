```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Standard_tag")
```

```python
df = pd.DataFrame(columns = ['ReviewNo','P','N','PP','NN','NP','PN'])
```

```python
for index, row in data.iterrows():
  if row['R'] not in df['ReviewNo']:
    df2 = {'ReviewNo': row['R'], 'P': 0, 'N': 0,'PP': 0 , 'NN': 0 , 'NP': 0, 'PN': 0}
    df = df.append(df2, ignore_index = True)
  
  rowNo = df.index[row['R']]

  if row['S'] == 'P':
    df.at[rowNo, 'P'] = df.at[rowNo, 'P'] +1
  
  if row['S'] == 'N':
    df.at[rowNo, 'N'] = df.at[rowNo, 'N'] +1

  if row['S'] == 'PP':
    df.at[rowNo, 'PP'] = df.at[rowNo, 'PP'] +1

  if row['S'] == 'PN':
    df.at[rowNo, 'PN'] = df.at[rowNo, 'PN'] +1

  if row['S'] == 'NN':
    df.at[rowNo, 'NN'] = df.at[rowNo, 'NN'] +1

  if row['S'] == 'NP':
     df.at[rowNo, 'NP'] = df.at[rowNo, 'NP'] +1

print(df)
   
df.to_excel("thirdstep_output.xlsx")
  
```
